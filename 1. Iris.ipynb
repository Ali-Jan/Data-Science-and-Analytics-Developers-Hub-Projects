```python
col = list(df.columns)
plt.figure(figsize=(5 * len(col), 4)) 

for idx, col_name in enumerate(col):
    plt.subplot(1,len(col),idx + 1)
    sns.histplot(df[col_name])
    
plt.tight_layout()
plt.show()
```
